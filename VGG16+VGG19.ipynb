create a function to append all the images and their labels in 'x' and 'y'

```python
x = []
y = []
img_size=(224,224)
def make_whole_dataset(directory,flower):
    try:
        for img in os.listdir(directory):
            path = os.path.join(directory,img)
            img_array = cv2.imread(path,cv2.IMREAD_COLOR)
            img_array = cv2.resize(img_array,img_size)
            x.append(np.array(img_array))
            y.append(str(flower))
    except:
        None

    print("flower {} has been added to Dataset Successfully".format(flower))
```

```python
make_whole_dataset(daisy_dir,"Daisy")
make_whole_dataset(rose_dir,"Rose")
make_whole_dataset(dandelion_dir,"Dandelion")
make_whole_dataset(sunflower_dir,"Sunflower")
make_whole_dataset(tulip_dir,"Tulip")
```

```python
#ONe hot encoding
le=LabelEncoder()
y=le.fit_transform(y)
y=to_categorical(y,5)
```

```python
x=np.array(x)
x=x/255.0
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=42)
```

Importing Vgg16 model

```python
vgg=VGG16(input_shape=[224,224,3],weights='imagenet',include_top=False)
```

```python
vgg.summary()
```

```python
for layers in vgg.layers[:16]:
    layers.trainable=False
```

> Add the flatten layer 

```python
from tensorflow.keras.layers import Input, Lambda, Dense, Flatten
x = Flatten()(vgg.output)
```

Add the dense layer

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense
prediction = Dense(5, activation='softmax')(x)
# create a model object
model = Model(inputs=vgg.input, outputs=prediction)
```

```python
model.summary()
```

Compile the model

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

fit the model without using data augmentation

```python
history=model.fit(train_x,train_y,epochs=25,validation_data=(test_x,test_y),batch_size=32)
```

```python
base_model = VGG19(weights='imagenet',include_top=False, input_shape=(224,224,3))
```

```python
base_model.summary()
```

```python
p=base_model.output
p=Flatten()(p)
p=Dense(500, activation='relu')(p)
p=Dropout(0.2)(p)
predictions = Dense(5, activation='softmax')(p)
model = Model(inputs=base_model.input, outputs=predictions)

```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```
